## 🧹 7. Data Cleaning
Here we execute structured data cleaning steps. For each step, we explain the engineering rationale and execute the code.

### Step 7.1: Dropping Irrelevant Features
**Why it is needed:** The columns `User ID`, `User Name`, `Driver Name`, `key`, and `pickup_datetime` contain unique identifiers, raw strings, and PII. They do not contain any generalized predictive signal.
**Risk of keeping:** Keeping them increases memory usage and causes tree-based models to overfit on noise (especially high-cardinality features like User ID).
**Action:** Drop them.


```python
# Drop irrelevant and high-cardinality columns
columns_to_drop = ['User ID', 'User Name', 'Driver Name', 'key', 'pickup_datetime'  , "ewr_dist", "sol_dist"]

df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
print(f'Columns remaining: {list(df.columns)}')
```

### Step 7.2: Dropping Missing Values
**Why it is needed:** There are 5 rows with null dropoff coordinates and distances.
**Action:** Drop these rows.


```python
# Drop rows with null values
print(f'Shape after dropping nulls: {df.shape}')
```

### Step 7.3: Validating Fare Amount
**Why it is needed:** In NYC, the minimum base taxi fare is $2.50 Negative fares and fares below $2.50 are invalid data.

```python
# Filter fare_amount
df = df[(df['fare_amount'] >= 2.5) ]
print(f'Shape after fare filtering: {df.shape}')
```

### Step 7.4: Validating Coordinates (NYC Bounding Box)
**Why it is needed:** NYC taxi rides must start and end within metropolitan limits. Coordinates outside this zone (e.g. 0,0) represent GPS glitches.
**Bounding Box limits (in degrees):**
- Longitude: $[-74.30, -73.65]$
- Latitude: $[40.45, 40.95]$
**Note on Radians:** The coordinates in the dataset are already stored in **radians**. Thus, we must convert our bounding box boundaries to radians before filtering.
**Action:** Apply the bounding box filter.


```python
# Define NYC bounding box boundaries in degrees and convert to radians
lon_min, lon_max = np.deg2rad(-74.30), np.deg2rad(-73.65)
lat_min, lat_max = np.deg2rad(40.45), np.deg2rad(40.95)

# Apply bounding box filter
df = df[
    df['pickup_longitude'].between(lon_min, lon_max) &
    df['pickup_latitude'].between(lat_min, lat_max) &
    df['dropoff_longitude'].between(lon_min, lon_max) &
    df['dropoff_latitude'].between(lat_min, lat_max)
]
print(f'Shape after coordinate filtering: {df.shape}')
```

### Step 7.5: Validating Passenger Count
**Why it is needed:** Passenger count must be between 1 and 4. Rides with 0 passengers are invalid. Passenger counts of 5 or 6 (UberXL) represent a separate pricing tier. To build a standard, reliable prediction model, we restrict passenger count to $[1, 4]$ and note that UberXL/SUV trips should be modeled separately.
**Action:** Restrict passenger_count to $[1, 4]$.


```python
# Filter passenger_count
df = df[(df['passenger_count'] >= 1) & (df['passenger_count'] <= 4)]
print(f'Shape after passenger count filtering: {df.shape}')
```

### Step 7.6: Validating Ride Distance
**Why it is needed:** Distance values above 100 km are extreme outliers for NYC metropolitan taxi trips. We also filter out 0 distance trips since they represent cancellations, waits, or GPS errors, rather than actual travel distance.
**Action:** Filter distance to $(0, 100]$ km.


```python
# Filter distance
logical_max_distance = 100.0
df = df[(df['distance'] > 0.0) & (df['distance'] <= logical_max_distance)]
print(f'Final cleaned dataset shape: {df.shape}')
```

## ⚙️ 8. Feature Engineering
Feature engineering is crucial for enabling algorithms (especially linear regression and neural networks) to learn complex patterns. The dataset already has calculated great-circle distance and airport distances, but we will add **Cyclic Temporal Encodings**.

### Cyclic Temporal Encoding
Temporal features like hour of the day (0-23), day of the week (0-6), and month of the year (1-12) are circular. For example, 23:00 (11 PM) is as close to 00:00 (Midnight) as 01:00 is to 02:00. However, in standard numerical representation, the distance between 23 and 0 is 23. This confuses distance-based models.

To resolve this, we map these features onto a 2D circle using sine and cosine transformations:
$$x_{sin} = \sin\left(\frac{2\pi \cdot x}{\text{Period}}\right)$$
$$x_{cos} = \cos\left(\frac{2\pi \cdot x}{\text{Period}}\right)$$

#### **Advantages:**
- Maps circular time sequences continuously, preserving proximity relationships (e.g. 23:00 close to 0:00).
- Crucial for Neural Networks, Support Vector Regressors, and Linear/Ridge Regression models.

#### **Disadvantages:**
- Adds dimensionality (doubles the number of temporal columns).
- Tree-based models (XGBoost, Random Forests) don't naturally benefit from this because they partition variables using splits, though it does not harm their performance.

Let's execute cyclic encoding for **hour**, **weekday**, and **month**.


```python
# Cyclic encoding for hour (period = 24)
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24.0)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24.0)

# Cyclic encoding for weekday (period = 7)
df['weekday_sin'] = np.sin(2 * np.pi * df['weekday'] / 7.0)
df['weekday_cos'] = np.cos(2 * np.pi * df['weekday'] / 7.0)

# Cyclic encoding for month (period = 12)
df['month_sin'] = np.sin(2 * np.pi * (df['month'] - 1) / 12.0)
df['month_cos'] = np.cos(2 * np.pi * (df['month'] - 1) / 12.0)

# Display engineered features
df[['hour', 'hour_sin', 'hour_cos', 'weekday', 'weekday_sin', 'weekday_cos']].head()
```

```python
df["is_weekend"] = df["weekday"].isin([5, 6]).astype("int8")

df["rush_hour"] = (
    ((df["hour"] >= 7) & (df["hour"] <= 9)) |
    ((df["hour"] >= 17) & (df["hour"] <= 22))
).astype("int8")

df["is_night"] = (
    (df["hour"] >= 23) | (df["hour"] <= 5)
).astype("int8")
```

```python
col_drop = ['hour' , 'day' , 'month' , 'weekday']

df.drop(columns=col_drop, inplace=True)
```

```python
df["Car Condition"] = df["Car Condition"].map({
    "Bad": 0,
    "Good": 1,
    "Very Good": 2,
    "Excellent": 3
})

df["Traffic Condition"] = df["Traffic Condition"].map({
    "Flow Traffic": 0,
    "Dense Traffic": 1,
    "Congested Traffic": 2
})
```

```python
x = df.drop('fare_amount' , axis= 1)
y = (df["fare_amount"])
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test  , y_train  , y_test = train_test_split( x , y , test_size= 0.2 , random_state= 5 )
```

```python
numeric_features = [
    "Car Condition",
    "Traffic Condition",
    "pickup_longitude",
    "pickup_latitude",
    "dropoff_longitude",
    "dropoff_latitude",
    "passenger_count",
    "year",
    "jfk_dist",

    "lga_dist",
    "nyc_dist",
    "distance",
    "bearing",
    "hour_sin",
    "hour_cos",
    "weekday_sin",
    "weekday_cos",
    "month_sin",
    "month_cos",
    "rush_hour",
    "is_weekend",
    "is_night",

]

categorical_features = [
    "Weather"
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ("num", RobustScaler(), numeric_features),
        ("weather", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(
    random_state=42,
    verbose=0,
    loss_function="RMSE",
    task_type="GPU",
    devices="0",
    bootstrap_type="Bernoulli"
)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", model)
])
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {

    "model__depth": [4,6,8,10],

    "model__learning_rate": [0.03,0.05,0.1],

    "model__iterations": [300,500,700],

    "model__l2_leaf_reg": [1,3,5,7,9],

    "model__subsample": [0.7,0.8,0.9,1.0]

}

random_search = RandomizedSearchCV(

    estimator=pipeline,

    param_distributions=param_dist,

    n_iter=10,

    scoring="neg_root_mean_squared_error",

    cv=5,

    random_state=42,

    n_jobs=1,

    verbose=2

)

random_search.fit(x_train, y_train)

print("Best Parameters:")
print(random_search.best_params_)

print("\nBest CV RMSE:")
print(-random_search.best_score_)
```

```python
best_model = random_search.best_estimator_
```

```python
pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

pred = best_model.predict(x_test)

print("MAE :", mean_absolute_error(y_test, pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, pred)))
print("R2 :", r2_score(y_test, pred))
```

```python
results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": pred
})

results["Error"] = abs(
    results["Actual"] - results["Predicted"]
)
```
